# Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import RandomOverSampler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("/content/drive/MyDrive/ClassroomML/bank-data/bank-full.csv", sep=';')
df.head()
```

```python
df.describe()
```

```python
X = df.iloc[:, :-1]
X
```

```python
y = df.iloc[:, -1]
y
```

```python
X_encoded = pd.get_dummies(X)
X_encoded
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=0)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled.shape, X_test_scaled
```

```python
oversampler = RandomOverSampler(sampling_strategy='minority')
X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train_scaled, y_train)
```

```python
X_train_resampled.shape, y_train_resampled.shape
```

```python
lreg_model = LogisticRegression()
```

```python
lreg_model.fit(X_train_resampled, y_train_resampled)
```

```python
pred = lreg_model.predict(X_test_scaled)
accuracy_score(y_test,pred)
```

```python
cm=confusion_matrix(y_test,pred)
labels=['yes','no']
plt.title('Confusion matrix')
sns.heatmap(cm,cbar=False,xticklabels=labels,yticklabels=labels,fmt='d',annot=True,cmap=plt.cm.Blues)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
cm
```

```python
#Calculating classification metrics
tp,fn,fp,tn=cm.ravel()
precision=tp/(tp+fp)*100
recall=tp/(tp+fn)*100
accuracy=(tp+tn)/(tp+tn+fp+fn)*100
F1score=(2*precision*recall/(precision+recall))

print("Precision:",precision)
print("Recall:",recall)
print("F1 score:",F1score)
print("Accuracy:",accuracy)
```
